###Model Fitting

```python
import statsmodels.formula.api as snf
```

```python
model = snf.ols("delivery_time~sorting_time",data=dataset).fit()
#Dependent Variable = Delivery Time and Independent variable is Sorting Time
```

###Model Testing

#2) Salary_hike -> Build a prediction model for Salary_hike

```python
import pandas as pd
dataset = pd.read_csv("Salary_Data.csv")
dataset.head()
```

###EDA

```python
sns.distplot(dataset['Salary'])
```

###Correlation Analysis


```python
dataset.corr()
```

Visualization of Correlation beteen x and y
regplot = regression plot


```python
sns.regplot(x=dataset['YearsExperience'],y=dataset['Salary'])
```

There is good correlation between the two variable.

The score is more than 0.8 which is a good sign

###Model Fitting

```python
model1=smf.ols("Salary~YearsExperience",data=dataset).fit()
```

###Model Testing

```python
model1.params

#coefficients Parameters
```

```python
#Pvalues and tvalues
model1.tvalues, model1.pvalues
```

```python
# Rsquared values
model1.rsquared , model1.rsquared_adj
```

```python
#To see the model predictions
model1.fittedvalues
```

Checking predictions of Model by equation
From the result Bo = 25792.200198668703 and B1 = 9449.962321455077, so our linear regression equation will become Y = Bo + B1X
Y = 25792.200198668703 + 9449.962321455077*X

```python
dataset['salary_Predicted'] = model1.fittedvalues
dataset[0:5]
```

```python
# Plotting Actual Vs. Predicted Values
plt.scatter(dataset.YearsExperience, dataset.Salary,  color='gray')
plt.plot(dataset.YearsExperience, dataset.salary_Predicted, color='red', linewidth=2)
plt.title('Actual Vs. Predicted Values')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
#To generate the residual values
#Difference between Actual & Fitted values
dataset['Error'] = model1.resid   #redid = residuals
dataset[0:5]
```

```python
dataset['Y'] = 25792.200199 + (9449.962321 + dataset['Salary'] + dataset['Error'])
dataset[0:5]
```

```python
dataset['Error_Sq'] = (dataset['Error'])**2
dataset[0:5]
```

```python
Mean_Squared_Error2 = dataset['Error_Sq'].mean()
Mean_Squared_Error2
```

```python
model1.rsquared
```

95.69% of variation in the dependent variable is explained by the given independent variable

```python
model1.summary()
```

The R-squared and Adjusted R-squared scores are above 0.85.

(It is a thumb rule to consider Adjusted R-squared to be greater than 0.8 for a good model for prediction)

F-statitics is quite high as well and yes desire it to be higher

But log-likelihood is quite very low far away from 0
and AIC and BIC score are much higher for this model
